```python
from underdogcowboy import Timeline, CommandProcessor, ModelManager, LLMConfigManager

# Initialize the timeline and model
timeline = Timeline()
config_manager = LLMConfigManager()
provider, model_name = config_manager.select_model()
model = ModelManager.initialize_model_with_id(provider, model_name)

# Create a CommandProcessor instance
processor = CommandProcessor(timeline, model)

# Sample system messages
system_messages = [
    "You are a helpful assistant.",
    "You are a pirate. Respond in pirate speak.",
    "You are a poet. Respond in rhymes.",
    "You are a scientist. Explain things in scientific terms.",
]

# Test each system message
for i, message in enumerate(system_messages, 1):
    print(f"\n--- Test {i}: {message} ---")
    
    # Set the system message
    timeline.set_system_message(message)
    
    # View the current system message
    system_msg = timeline.get_system_message()
    print(f"Current system message: {system_msg.text}")

    # Process a sample user message
    user_input = "Tell me about the weather today."
    print(f"\nUser: {user_input}")
    
    response = processor.process_single_message(user_input)
    print(f"Assistant: {response}")

    # Clear the timeline for the next test
    timeline.history.clear()

# Test deleting the system message
print("\n--- Test: Deleting System Message ---")
timeline.delete_system_message()
system_msg = timeline.get_system_message()
print(f"System message after deletion: {system_msg}")

# Process a message without a system message
user_input = "What's your role now?"
print(f"\nUser: {user_input}")
response = processor.process_single_message(user_input)
print(f"Assistant: {response}")
```
